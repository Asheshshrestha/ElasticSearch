```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
import xlwt
from xlwt import Workbook
wb = Workbook()
sheet = wb.add_sheet('Sheet 16')
sheet.write(0,0,"S/N")
sheet.write(0,1,"Movie Name")
sheet.write(0,2,"Rating")
sheet.write(0,3,"Duration")
sheet.write(0,4,"Type")
sheet.write(0,5,"Release Year")
sheet.write(0,6,"Director")
sheet.write(0,7,"Stars")
sheet.write(0,8,"Descreption")

```

```python
total_movie=10000  #insert the total movie you want to extract
count=0
row = 0
for page in range(140451,2000000,50): #(starting_movie_value,Last_movie_value)
    if count>= total_movie:
                break
    my_url = 'https://www.imdb.com/search/title/?genres=action&start='+str(page)+'&explore=title_type,genres&ref_=adv_nxt'
    
    uClient = uReq(my_url)
    page_html = uClient.read()
    uClient.close()


    page_soup = soup(page_html,"html.parser")
    #print(page_soup)

    cards = page_soup.findAll("div",{"class":"lister-item-content"})
    #print(len(cards))

    for movie in cards:
        try:
            count = count+1
            #print(count)
            row = row+1
            #print(row)
            sheet.write(row,0,count)
            #print("laa")
            name = movie.h3.a.text
            #print("Name: "+ name )
            sheet.write(row,1,name)
            #print(name)
            try:
                if movie.div != None:
                    rating = movie.div.div.strong.text
                    #print("Rating:"+ rating)
                else:
                    rating = "N/A"
                    #print("Rating:"+rating)
            except:
                rating = "N/A"
                #print("Rating:"+rating)
            sheet.write(row,2,rating)
            try:   
                if movie.p.find("span",{"class":"runtime"}) != None:
                    strtest1 = movie.p.find("span",{"class":"runtime"}).text
                    strtest1 = strtest1.replace('\n','').replace('\r','')
                    duration = strtest1[:-4]
                    #print("Duration:"+duration+" min")
                else:
                    duration = "N/A"
                    
                    #print("Duration :"+ duration)
            except:
                duration = "N/A"
                #print("Duration :"+ duration)
            sheet.write(row,3,duration)
            try: 
                strtest = movie.p.find("span",{"class":"genre"}).text
                strtest = strtest.replace('\n','').replace('\r','')
                _type = strtest
                #print("Type: "+_type)
            except:
                _type = "N/A"
                #print("Type: "+_type)
            sheet.write(row,4,_type)
                
            try:
                release_year = movie.h3.find("span",{"class":"lister-item-year"}).text[1:-1]
                
                if release_year[-2] == "–":
                    release_year = release_year[:-2]
            except:
                release_year="N/A"
            sheet.write(row,5,release_year)
            #print("Release Year: "+ release_year)
            try:
                director = movie.find_all("p")[2].text.strip()
                if director[0] == "D":
                    director = movie.find_all("p")[2].find_all("a")[0].text
                    #print("Director :" + director )
                else:
                    director = "N/A"
                    #print("Director :"+director)
            except:
                director = "N/A"
                #print("Director :"+director)
            sheet.write(row,6,director)
            try:
                stars = movie.find_all("p")[2].text.strip()
                
                if stars[0] == "S":
                    stars = movie.find_all("p")[2].find_all("a")[0].text+","+movie.find_all("p")[2].find_all("a")[1].text+","+movie.find_all("p")[2].find_all("a")[2].text+","+movie.find_all("p")[2].find_all("a")[3].text
                    #print("Stars: "+ stars)
                elif stars[0] == "D":
                    index = stars.find('|')
                    stars = stars[index+1:].strip()
                    stars = stars.replace('\n','').replace('\r','')
                    stars = stars[6:]
                    #print("Stars:" + stars)
                    
            except:
                stars = "N/A"
                #print("Stars:"+stars)
            sheet.write(row,7,stars)
            try:
                discription = movie.find_all("p",{"class":"text-muted"})[1].text[5:].replace('\n','').replace('\r','')
                #print("Descreption:" + discription)
            except:
                discription = "N/A"
                #print("Descreption:" + discription)
            sheet.write(row,8,discription)
                
            #print("***"*41)
            
        except:
            print("ex")
            count = count-1
            row = row - 1
        finally:
            #print(count)
            wb.save('test15.xls')
            if count>= total_movie:
                break

wb.save('test15.xls')
print("End of the session")
```
